# 2)Average number of words in movie titles between the year 2000-2005

```python
df_title = df[(df['release_year']>1999) & (df['release_year']<2006)]
```

```python
df_title
```

```python
word_count = []
for i in range(df_title.shape[0]):
    current_title = str(df['original_title'].values[i])
    current_title_list = current_title.split(' ')
    n=0
    for each_word in current_title_list:
        n = n+1
    word_count.append(n)
```

```python
average_list = sum(word_count) / len(word_count)
```

# 3) Most common Genre for Vin Diesel & Emma Watson movies

```python
df.dropna(axis="rows",how="any",subset=["cast"],inplace=True)
```

```python
list_vin_diesel = []
list_emma_watson = []
for i in range(df.shape[0]):
    cast = str(df['cast'].values[i])
    cast_list = cast.split('|')
    for member in cast_list:
        if member == 'Vin Diesel':
            genre = str(df['genres'].values[i])
            genre_list = genre.split('|')
            for type in genre_list:
                list_vin_diesel.append(type)
        elif member == 'Emma Watson':
            genre = str(df['genres'].values[i])
            genre_list = genre.split('|')
            for type in genre_list:
                list_emma_watson.append(type)
```

```python
df_vin_diesel = pd.DataFrame(list_vin_diesel, columns =['Genre'])
df_emma_watson = pd.DataFrame(list_emma_watson, columns =['Genre'])
```
